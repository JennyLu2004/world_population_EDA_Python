```python
# Import the csv file
df = pd.read_csv("world_population.csv")
df
```

```python
# Change the float format of number columns to 2 decimals 
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
# Getting unique values for each columns
df.nunique()
```

```python
# Sort data frame by 2022 Population and get the top 10 of the most populated country in 2022
df.sort_values('2022 Population', ascending = False).head(10)
```

```python
# Looking at correlation between numeric columns
df.corr(numeric_only = True)
# We can see the correlation in population among the years are strongly correlated. Yet, the population are only somewhat correlated with the area 
```

```python
# Create the heatmap of correlations
sns.heatmap(df.corr(numeric_only = True), annot = True)
```

```python
# Create a list of columns of population 
population = ['1970 Population',
       '1980 Population', '1990 Population', '2000 Population',
       '2010 Population', '2015 Population', '2020 Population',
       '2022 Population']

# Group by continent and get the mean values from each column and sort by descending order
df2 = df.groupby('Continent')[population].mean(numeric_only = True).sort_values(by = "2022 Population", ascending = False)
df2
```

```python
df.columns
```

```python
# Flipping the columns and rows (y axis and x axis)
df3 = df2.transpose()
```

```python
# Plot continent and population by year
df2.plot()
# This is not a good graph, we need to switch the year to the x axis and vice versa
```

```python
# plot again with flipped axes
df3.plot()
```
